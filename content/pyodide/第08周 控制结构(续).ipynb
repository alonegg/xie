````python
from random import random
from math import sqrt
darts = 10000
hits = 0
for i in range(1, darts + 1):
    x, y = random(), random()
    dist = sqrt(x ** 2 + y ** 2)
    if dist <= 1.0:
        hits = hits + 1
pi = 4 * (hits/darts)
print("Pi{}".format(pi))
````

````python
loan_amt = float(input(':'))
loan_month = int(input('():'))
month_rate = 0.0435 / 12 
repaid_amt = 0 
loan_amt_per_month = loan_amt / loan_month 
for i in range(1, loan_month + 1):
    interest_amt = (loan_amt - repaid_amt) * month_rate 
    amt_for_month = loan_amt_per_month + interest_amt 
    repaid_amt = repaid_amt + loan_amt_per_month 
    print('{}:{:.2f}, :{:.2f}, :{:.2f}:{:.2f}'.format(i,
                                            amt_for_month,loan_amt_per_month,
                                            interest_amt,loan_amt - repaid_amt))
````

````python
name_list = []
amount_list = []
print('')
print('-' * 65)
i = 1
while True:
    tmp_str = input('{}'.format(i))
    if len(tmp_str) == 0:
        break
    name, amount = tmp_str.split(',')
    name_list.append(name)
    amount_list.append(float(amount))
    i += 1
print('-' * 65)
search_name = input('')
search_idx = name_list.index(search_name)
total_amount = sum(amount_list)
search_amount = amount_list[search_idx]
print('-' * 65)
print('{}{:.2f}{:.2%}'.format(search_name,
                                          search_amount, search_amount / total_amount))
````
