## 2.  random库应用： $\pi$的计算

$\pi$（圆周率）是一个无理数，即无限不循环小数。精确求解圆周率$\pi$是几何学、物理学和很多工程学科的关键。

对$\pi$的精确求解曾经是数学历史上一直难以解决的问题之一，因为$\pi$无法用任何精确公式表示，在电子计算机出现以前，$\pi$只能通过一些近似公式的求解得到，直到1948年，人类才以人工计算方式得到$\pi$的808位精确小数。

随着计算机的出现，数学家找到了另类求解π的另类方法：***蒙特卡罗（Monte Carlo）方法***，又称随机抽样或统计试验方法。当所要求解的问题是某种事件出现的概率，或者是某个随机变量的期望值时，它们可以通过某种“试验”的方法，得到这种事件出现的频率，或者这个随机变数的平均值，并用它们作为问题的解。这就是蒙特卡罗方法的基本思想。

应用蒙特卡罗方法求解$\pi$的基本步骤如下：<br>
* 随机向单位正方形和圆结构，抛洒大量“飞镖”点
* 计算每个点到圆心的距离从而判断该点在圆内或者圆外
* 用圆内的点数除以总点数就是$\pi$/4值。

随机点数量越大，越充分覆盖整个图形，计算得到的$\pi$值越精确。实际上，这个方法的思想是利用离散点值表示图形的面积，通过面积比例来求解$\pi$值。 



````python
from random import random
from math import sqrt
darts = 10000
hits = 0
for i in range(1, darts + 1):
    x, y = random(), random()
    dist = sqrt(x ** 2 + y ** 2)
    if dist <= 1.0:
        hits = hits + 1
pi = 4 * (hits/darts)
print("Pi值是{}".format(pi))
````

## 3. 列表推导式
推导式是从可迭代对象中快速简洁的创建数据类型的一种方法。它使得你可以用优美简短的代码就能实现循环甚至条件判断。使用推导式的代码被认为更像Python代码，或者说这样的代码更Python。列表推导式会产生一个新的列表，其语法形式如下所示：
```python
[expr for item in iterable_obj]
```

````python
[item ** 2 for item in range(0,11)]
````

````python
s = []
for i in range(0,11):
    if i%2==0:
        s.append(i ** 2)
s
````

#### <font color=#A93226>【课堂练习】</font> 

<font size=3 face="楷体-简">从一个字符串s1中返回另一个字符串s2的全部索引。例如，对于字符串`s1='EACBDCEACCAC'`,`s2='AC'`，返回的结果为：`[1, 7, 10]`。</font>


````python
#方法1：常规
import random,string
s1="".join([random.choice(string.ascii_uppercase[:5]) for i in range(10)])
print(s1)
s2="A"
res=[]
pos=0
while True:
    w=s1.find(s2,pos)
    if w<0:
        break
    res.append(w)
    pos=w+1
print(res)
````

````python
#方法2:这个方法只能判断单个字符，如果s2为长度大于1的字符串呢？
import random,string
s1="".join([random.choice(string.ascii_uppercase[:5]) for i in range(10)])
print(s1)
s2="AB"
res=[idx for idx,b in enumerate(s1) if b==s2]
print(res)
````

````python
#方法3：解决方法2中的问题
import random,string
s1="".join([random.choice(string.ascii_uppercase[:5]) for i in range(20)])
print(s1)
s2="AB"
res=[i for i in range(0,len(s1)-len(s2)) if s1[i:i+len(s2)]==s2]
print(res)
````

#### <font color=#A93226>【课堂练习】</font> 

<font size=3 face="楷体-简">古代中国使用天干地支来记录当前的年份。<br/><br/>

天干一共有10个，分别为：甲(jiǎ)、乙(yi)、丙 (bǐng)、丁(ding)、戊(wù)、已(j)、庚(géng)、辛(xin)、壬(rén)、癸(guǐ)。<br/>
地支一共有12个，分别为：子(zǐ)、丑(chǒu)、寅(yín)、卯(mǎo)、辰(chén)、巳(si)、午(wǔ)、未(wèi)、申(shēn)、酉(yǒu)、戌(xū)、亥(hài)。<br/>
    
将天干和地支连起来，就组成了一个天干地支的年份，例如：甲子，第一个甲子年是汉安帝延光三年甲子岁（公历124年）。2020年是庚子年，每过一年，天干和地支都会移动到下一个。例如 2021年是辛丑年。每过 60 年，天干会循环6轮，地支会循环5轮，所以天干地支纪年每60年轮回一次。例如 1900年，1960年，2020年都是庚子年。
给定一个公元纪年的年份，请输出这一年的天干地支年份。</font>

#### 输入格式
输入一行包含一个正整数，表示公元年份。例如：2024
#### 输出格式
输出表示天干地支的年份。例如：2024年是甲辰年

#### 解法1：利用最大公约数和字符串连接

````python
tian = '甲乙丙丁戊已庚辛壬癸'
di   = '子丑寅卯辰巳午未申酉戌亥'

len_tian, len_di = len(tian), len(di)

a = max(len_tian, len_di)
while a< len_tian*len_di:
    if a%len_tian==0 and a%len_di==0:
        break
    a += 1

tian_more = tian * (a//len_tian)
di_more = di * (a//len_di)

year = int(input("请输入年份:"))
i = year%124
print(f"{year}年是{tian_more[i] + di_more[i]}年")
````

#### 解法2：利用最大公约数和zip函数

````python
tian = '甲乙丙丁戊已庚辛壬癸'
di   = '子丑寅卯辰巳午未申酉戌亥'

len_tian, len_di = len(tian), len(di)
a = max(len_tian, len_di)
while a< len_tian*len_di:
    if a%len_tian==0 and a%len_di==0:
        break
    a += 1
    
tian_di = zip(tian * (a//len_tian), di * (a//len_di))

year = int(input("请输入年份:"))
i = year%124
print(f"{year}年是{''.join(list(tian_di)[i])}年")
````

#### 解法3：利用余数

````python
tian = '甲乙丙丁戊已庚辛壬癸'
di   = '子丑寅卯辰巳午未申酉戌亥'
len_tian, len_di = len(tian), len(di)
year = int(input("请输入年份:"))
i, j = (year-124)%len_tian, (year-124)%len_di
print(f"{year}年是{tian[i]+di[j]}年")
````

<a href="https://www.xuetangx.com/learn/swufe08091006588/swufe08091006588/19321427/video/43239981" style="color:#BDC3C7; text-decoration:none; font-size: 10pt;">《观看在线视频》</a>

<font size=3 face="楷体-简">贷款的还款方式分为等额本息还款和等额本金还款。等额本息还款是指借款人每月以相等的金额偿还贷款本息。等额本金还款是在还款期内把贷款数总额等分，每月偿还同等 数额的本金和剩余贷款在该月所产生的利息，这样由于每月的还款本金额固定，而利息越来越少，借款人起初还款压力较大，但是随时间的推移每月还款数越来越少。

等额本息还款所还的利息高，但前期还款压力不大，适合一般的工薪族。等额本金还款所还的利息低，但前期还款压力大，适合经济收入较好的家庭。

编写程序，根据用户输入的贷款金额和贷款期限， 假设年利率为 4.35%， 利用等额本金还款方式，计算每个月的还款金额（包含的本金、利息）及剩余金额。
</font>

````python
loan_amt = float(input('请输入贷款金额:'))
loan_month = int(input('请输入贷款期限(月):'))
month_rate = 0.0435 / 12 
repaid_amt = 0 
loan_amt_per_month = loan_amt / loan_month 
for i in range(1, loan_month + 1):
    interest_amt = (loan_amt - repaid_amt) * month_rate 
    amt_for_month = loan_amt_per_month + interest_amt 
    repaid_amt = repaid_amt + loan_amt_per_month 
    print('第{}个月须还:{:.2f}, 其中本金:{:.2f}, 利息:{:.2f}。本金余额:{:.2f}'.format(i,
                                            amt_for_month,loan_amt_per_month,
                                            interest_amt,loan_amt - repaid_amt))
````

#### <font color=#A93226>【课堂练习】</font> 

Aya 定义符合如下条件的正整数`x`为优秀正整数,
* `x`为一个完全平方数;
* `x`的各位数和为质数。

例如：25是一个优秀正整数，因为`25=5*5`，为完全平方数，且`2+5=7`，为质数，
现在给定L,R，请求出L 到 R(包含L和R)的正整数中，所有优秀正整数的乘积，对 998244353 取模的值。若不存在优秀正整数，则输出 0。

````python
num1, num2 = map(int, input("请输入两个数，用逗号隔开：").split(","))

result = 1
for x in range(num1, num2+1):
    # 判断x是否为完全平方数
    if x!=int(x**0.5)**2:
        continue
        
    # 求各位数的和
    y = sum(map(int, str(x)))
    for m in range(2, int(y**0.5)+1):
        if y%m==0:
            break
    else:
        # 各位数和为质数，所以x为优秀正整数
        result *= x
        
if result==1:
    print(0)
else:
    print(result%998244353)
````

## 5. 引导案例解析
根据对引导案例的分析，结合本章的学习，利用while循环结构和if分支结构对引导案例改进，使得程序支持任意项明细情况的输入。
程序代码如下：

````python
name_list = []
amount_list = []
print('请输入学习用品支出情况，名称和金额用逗号分隔，直接回车结束输入。')
print('-' * 65)
i = 1
while True:
    tmp_str = input('第{}项：'.format(i))
    if len(tmp_str) == 0:
        break
    name, amount = tmp_str.split(',')
    name_list.append(name)
    amount_list.append(float(amount))
    i += 1
print('-' * 65)
search_name = input('请输入需要查询的学习用品种类名称：')
search_idx = name_list.index(search_name)
total_amount = sum(amount_list)
search_amount = amount_list[search_idx]
print('-' * 65)
print('学习用品中{}支出了{:.2f}元，占比：{:.2%}'.format(search_name,
                                          search_amount, search_amount / total_amount))
````
