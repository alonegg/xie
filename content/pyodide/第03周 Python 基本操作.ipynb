````python
# 法一
(3678.7 - 1470.1)/(1978-1960)
````

````python
# 法二,利用变量
start = 1470.1
end = 3678.7
timespan = 1978-1960
(end - start) / timespan
````

````python
(1015986.2 - 3678.7)/(2020-1978)
````

## 2.3.4 <font color=#A93226>案例：</font>股票价格增长率与涨跌幅的计算

<font size=3 face="楷体-简">增长率（growth rate）也称增长速度， 是指一定时期内某一数据指标的增长量与基期 数据的比值。 例如， 招商银行在 2002 年 4 月 9 日上市当日的收盘价为 2.59 元， 至 2020 年 12 月 31 日收盘价为 44.35 元，其收盘价折线图如图所示。</font>


<font size=3 face="楷体-简">在计算招商银行上市以来至 2020 年 12 月月底的收盘价增长率时， 可以将 2002 年 4 月 9 日的收盘价 2.59 元作为该时期期初值， 将 2020 年 12 月 31 日收盘价 44.35 元作 为该时期期末值， 用期末值除以期初值再减去 1， 得到招商银行股票价格在该段时期内的 增长率：</font>

````python
stock_beginning_price = 2.59
stock_ending_price = 44.35
(stock_ending_price / stock_beginning_price) - 1
````

<font size=3 face="楷体-简">涨跌幅是对涨跌值的描述， 用 % 表示， 涨跌幅 = 涨跌值 / 昨收盘 * 100%。 其中， 涨 跌值是当前交易日最新成交价（或收盘价）与前一交易日收盘价相减所产生的数值。 因 此，当日最新成交价比前一交易日收盘价高，涨跌幅为正，当日最新成交价比前一交易日收盘价低，则涨跌幅为负。

例如， 招商银行 2020 年 12 月 30 日收盘价为 43.05 元，2020 年 12 月 31 日收盘价为 44.35 元，所以，招商银行 2020 年 12 月 31 日的涨跌幅为：</font>

````python
price_20201230 = 43.05
price_20201231 = 44.35
(price_20201231 - price_20201230) / price_20201230
````

### 2.4.5 字符串format()方法<br>
字符串``format()``方法的基本使用格式是：<br>

```python
'…{<参数序号>}…'.format(<逗号分隔的参数列表>)
```
    
在模版字符串'…{}…'中的花括号相对位置（例如{}）或指定位置（例如{1}）来制定替换目标及要插入的参数。

<font color=#A93226><b>【例 2-14】</b></font> <font size=3 face="楷体-简">利用 format 方法， 将股票代码“600036”和股票名称“招商银行”组合成一个完整的字符串。</font>

````python
'股票代码:{} 对应的名称是：{}'.format('600036','招商银行')
````

````python
'股票代码:{1} 对应的名称是：{0}'.format('招商银行','600036')
````


在```format()```方法中模板字符串的槽除了包括参数序号，还可以通过冒号:分隔后，包括格式控制信息。此时，槽的内部样式如下：

```python
{<参数序号>: <格式控制标记>}
```
其中，格式控制标记用来控制参数显示时的格式 。格式控制标记包括：```<填充><对齐><宽度>,<.精度><类型>```个字段，这些字段都是可选的，可以组合使用，这里按照使用方式逐一介绍:

|:|<填充>|<对齐>|<宽度>|,|<.精度>|<类型>|
|:--:|:--:|:--:|:--:|:--:|:--:|:--:|
|引导<br>符号|用于填充的<br>单个字符|< 左对齐<br>> 右对齐<br>^ 居中对齐|槽的设定<br>输出宽度|数字的<br>千位分隔符<br>只用于整数和浮点数|浮点数<br>小数部分的精度<br>或<br>字符串<br>的最大输出长度|整数类型<br>b,c,d,o,x,X<br>浮点数类型<br>e,E,f,%|

````python
print('{:.<10}{:.>20.2f}'.format("宫保鸡丁",100.12345))
print('{:.<10}{:.>20.2f}'.format("宫保鸡丁鸡丁",10.12345))
print('{:.<10}{:.>20.2f}'.format("宫保鸡丁",1000.12345))
print('{:.<10}{:.>20.2f}'.format("宫保",108.12345))
````

````python
s = "Python"
"{0:30}".format(s)
````

````python
"{0:>30}".format(s)
````

````python
"{0:*^30}".format(s)
````

````python
"{0:.2f}".format(12345.67890)
````

<font color=#A93226><b>【例 2-14】</b></font> <font size=3 face="楷体-简">利用 format 方法， 将股票代码“600036”， 股票名称“招商银行”， 收盘价 36.84 元，涨跌幅 –1.15%，总市值 9291 亿元组合成一个完整字符串。</font>

#### f-字符串,一种新的增强型字符串格式化方式 ####
f-string，即格式化字符串常量（formatted string literals），Python3.6新引入的一种字符串格式化方法，主要目的是使格式化字符串的操作更加简便。

f-string在形式上是以 f 或 F 修饰符引领的字符串（f'xxx' 或 F'xxx'），以大括号 {} 标明被替换的字段。

f-string在本质上并不是字符串常量，而是一个在运行时运算求值的表达式。

#### 简单使用
f-string用大括号 {} 表示被替换字段，其中直接填入替换内容：

````python
year = 2022
month = 10
day = 16
name = '中国共产党第二十次全国代表大会'
f'{year}年{month}月{day}日上午，{name}在人民大会堂开幕。'
````

#### 表达式求值与函数调用
f-string的大括号 {} 可以填入表达式或调用函数，Python会求出其结果并填入返回的字符串内：

````python
pingshi = 95
qimo = 90
f'总分数是：{95 * 0.3 + 90 * 0.7}'
````

````python
university = 'Southwestern University of Finance and Economics'
f'{university.upper()}'
````

#### 自定义格式
f-string采用 ```{content:format}``` 设置字符串格式，其中 content 是替换并填入字符串的内容，可以是变量、表达式或函数等，format 是格式描述符。采用默认格式时不必指定 ``{:format}``，如上面例子所示只写 ``{content}`` 即可。

关于格式描述符的详细语法及含义可查阅[Python官方文档](https://docs.python.org/3/library/string.html#format-string-syntax)。

````python
print(f'{"宫保鸡丁":.<10}{100.12345:.>20.2f}')
print(f'{"宫保鸡丁鸡丁":.<10}{10.12345:.>20.2f}')
print(f'{"宫保鸡丁":.<10}{1000.12345:.>20.2f}')
print(f'{"宫保":.<10}{108.12345:.>20.2f}')
````
