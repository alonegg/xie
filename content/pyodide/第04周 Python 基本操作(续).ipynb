#### (2) 字符串方法upper()、lower()、isupper()和islower() <br>
``upper()``和``lower()``字符串方法返回一个新字符串，其中原字符串的所有字母都被相应地转换为大写或小写。字符串中非字母字符保持不变。

````python
sayHello = 'Hello world!'
print(sayHello.upper())
print(sayHello)   # 这些方法没有改变字符串本身，而是返回一个新字符串
````

````python
sayHello = sayHello.lower()
print(sayHello)
````

如果需要进行大小写无关的比较，``upper()``和``lower()``方法就很有用。例如，对于用户输入的名字，忽略大小写：

````python
userName = input('请输入用户名：')
userName.lower() == 'xzl'
````

如果字符串至少有一个字母，并且所有字母都是大写或小写，``isupper()``和``islower()``方法就会相应地返回布尔值True。否则，该方法返回False。

````python
'Hello world!'.isupper()
````

````python
'Hello world!'.islower()
````

````python
'Hello world!'.upper().isupper()
````

````python
'Hello world!'.upper().lower().isupper()  # 这样做的表达式看起来就像方法调用链。
````

````python
'Hello'.center(20, '=')
````

````python
leftWidth = 20
rightWidth = 6
print("Items".center(leftWidth + rightWidth, '-'))
chapter1 = "sandwiches"
chapter2 = "apples"
chapter3 = "cookies"
chp1Page = 30
chp2Page = 330
chp3Page =6100
print(chapter1.ljust(leftWidth, '.') + str(chp1Page).rjust(rightWidth))
print(chapter2.ljust(leftWidth, '.') + str(chp2Page).rjust(rightWidth))
print(chapter3.ljust(leftWidth, '.') + str(chp3Page).rjust(rightWidth))
````

#### (7) 用strip()、rstrip()和lstrip()删除空白字符 <br>
有时候你希望删除字符串左边、右边或两边的空白字符（空格、制表符和换行符）。
* ``strip()``字符串方法将返回一个新的字符串，它的开头或末尾都没有空白字符。
* ``lstrip()``方法将相应删除左边的空白字符。

````python
sayHello = ' Hello World '
print('-' + sayHello.strip() + '-')
print('-' + sayHello.lstrip() + '-')
print('-' + sayHello.rstrip() + '-')
print('-' + sayHello + '-')
````

## 2.4.9 <font color=#A93226>案例：</font>等额本息还款法每月还款额的计算

<a href="https://www.xuetangx.com/learn/swufe08091006588/swufe08091006588/23899599/video/58451022" style="color:#BDC3C7; text-decoration:none; font-size: 10pt;">《观看在线视频》</a>

<font size=3 face="楷体-简">等额本息还款法（等额法）， 即把按揭贷款的本金总额与利息总额相加， 平均分摊到 还款期限的每个月当中，每个月还款额是固定的，这样每个月还款额中的本金比重逐月递 增，利息比重逐月递减。

依据等额本息还款法的规则，可以推导出以下公式： 
$$ result = \frac{total * rate\_month * (1 + rate\_month)^{terms}}{(1 + rate\_month)^{terms} –1}$$

其中， result 为每月还款额， total 为贷款总额， rate_month 为银行月利率， terms 为 贷款总期数（按月算）。

编写程序，根据用户输入的贷款额度和年限，计算其利用等额本息还款法时每月的还款额。
</font>

````python
total = float(input("请输入贷款金额（元）："))
terms = int(input("请输入贷款期限（月）："))
rateYear = 0.049    #年利率为4.9%
rateMonth = rateYear / 12   #计算月利率
result = (total * rateMonth * (1 + rateMonth) ** terms) \
            / ((1 + rateMonth) ** terms - 1)
print("每月应还：",result)
````

### 2.5.2　逻辑运算
Python中3个布尔操作符（and、or 和not）用于计算布尔值。像比较操作符一样，它们将这些表达式求值为一个布尔值。

and和or操作符总是接受两个布尔值（或表达式），所以它们被认为是“**二元**”操作符。

**and**:如果两个布尔值都为True，and操作符就将表达式求值为True，否则求值为False。

````python
True and True
````

````python
True and False
````

````python
(1 == 1) and (10 > 100)
````

**or**:只要有一个布尔值为真，or 操作符就将表达式求值为True。如果都是False，所求值为False。

````python
False or True
````

````python
False or False
````

````python
(1 == 1) or (10 > 100)
````

**not**:和and 和or 不同，not 操作符只作用于一个布尔值（或表达式）。not 操作符求值为相反的布尔值。

````python
not True
````

````python
not not not not False
````

## 2.7 导入模块
在利用Python处理财经数据时，除了加减乘除等基本算术运算外，还会涉及到更加复杂的处理、运算和统计分析功能，这时可以使用Python的扩展工具：**模块**。

模块是第三方专门为了解决某些特定问题而编写的工具。Python本身自带了一些常用的模块，例如，math模块中具有较为复杂的求解正弦、余弦和平方根等运算，这些模块不需要安装，但是在使用前需要导入。

导入模块使用**import**关键字，Python中导入模块有不同的方法：

### （1）利用``import 模块名``
导入整个模块：

````python
import math
math.sqrt(4)
````

### （2）利用``import模块名 as 名称缩写``
导入整个模块的同时给该模块取个较短的别名：

````python
import numpy as np
np.sqrt(4)
````

### （3） 利用``import模块名.子模块名 as 名称缩写``
导入某个模块子模块的同时给该子模块取个较短的别名：

````python
import matplotlib.pyplot as plt
````

### （4）利用``from 模块名 import 函数``
导入模块中指定函数：

````python
from math import sqrt, exp
sqrt(4)
````

### （5）利用``from 模块名.子模块名 import 函数``
导入某个模块的子模块中指定函数：

````python
from matplotlib.pyplot import plot
````

我国人口数量一直保持世界人口较大比例。1850年我国人口数量约4.3亿，占世界人口的34%。1949年，我国大陆人口为5.4167亿，占世界人口比例为22%。由于人口结构的原因，我国人口总量仍在持续增长，下表是我国自1949年至2019年每年年末人口数：

根据上表，可画出如图2-3所示的1949年至2019年期间每隔10年的人口柱状图:

````python
# 这段代码暂时不能理解没有关系，继续往后面学习
import matplotlib.pyplot as plt
%matplotlib inline
years = [1949, 1959, 1969, 1979, 1989, 1999,2009, 2019]
population = [54167, 67207, 80671, 97542, 112704, 125786, 133450, 140005]
plt.bar(years, population , tick_label=years);
````

基于上述结果，计算自1949年至2019年期间以10年为间隔的人口增长率：

````python
rate1 = (67207 - 54167) / 54167
rate2 = (80671 - 67207) / 67207
rate3 = (97542 - 80671) / 80671
rate4 = (112704 - 97542) / 97542
rate5 = (125786 - 112704) / 112704
rate6 = (133450 - 125786) / 125786
rate7 = (140005 - 133450) / 133450
years_growth_rate = [1959,1969,1979,1989,1999,2009,2019]
pop_growth_rate = [rate1,rate2,rate3,rate4,rate5,rate6,rate7]
plt.xticks(years_growth_rate)
plt.plot(years_growth_rate, pop_growth_rate);
````

<font color=#A93226><b>【课堂案例】</b></font>  <font size=3 face="楷体-简">

读取Data/Sales目录下的销售数据文件，并打印出退货数量大于0的销售记录的销售总额。</font><br><br>

<font size=1 color="gray">
注：在这个案例中，将提前了解一些关于列表、循环和文件的内容，不能掌握也没有关系，后面章节会详细讲解。</font>

````python
import os
#获取目录下的文件名列表
fileNameList = [item for item in os.listdir('Data/Sales/') if item.endswith('.csv') and not item.startswith('.')]
saleAmt = 0
for fileName in fileNameList:  #循环处理每一个文件
    saleFile = open('Data/Sales/' + fileName, encoding='utf8')
    saleFile.readline()  #读取文件的第一行，即标题行
    for line in saleFile:  #从文件的第二行，即数据行开始处理每一行
        detail = line.replace('"', '').strip('\n').split(',')
        if int(detail[5])>0:  #只统计退货数量大于0的销售记录
            saleAmt = saleAmt + (int(detail[4]) - int(detail[5])) * float(detail[3]) 
print('销售总额为：{:.2f}元'.format(saleAmt))
````

<font color=#A93226><b>【课堂练习】</b></font>  <font size=3 face="楷体-简">

读取Data/Sales目录下的销售数据文件，并打印出所有商品的平均价格。</font>


````python
import os
fileNameList = [item for item in os.listdir('Data/Sales/') if item.endswith('.csv') and item.startswith('2020')]
price_sum = 0
qty = 0
for fileName in fileNameList:  
    saleFile = open('Data/Sales/' + fileName, encoding='utf8')
    saleFile.readline()  
    for line in saleFile:  
        detail = line.replace('"', '').strip('\n').split(',')
        price_sum += float(detail[3])
        qty += 1
                           
print('所有商品的平均价格为：{:.2f}元'.format(price_sum/qty))
````

## 【引导案例解析】
根据对引导案例的分析，首先需要输入姓名、日常生活支出、学习用品支出和其他方面支出等基本信息；接着计算出该月总支出；最后打印出各类支出的占比详情。

程序代码如下：

````python
stu_name = input('请输入姓名：')
living_expenses = float(input('请输入日常生活支出（元）：'))
school_things = float(input('请输入学习用品支出（元）：'))
other_expendses = float(input('其他方面支出（元）：'))
total_expenses = living_expenses + school_things + other_expendses
print('=' * 35)
print('以下是{}的支出情况：'.format(stu_name))
print('日常生活支出：{0:.2f}元，占比：{1:.2%}'.format(living_expenses, 
                                          living_expenses / total_expenses))
print('学习用品支出：{0:.2f}元，占比：{1:.2%}'.format(school_things, 
                                          school_things / total_expenses))
print('其他方面支出：{0:.2f}元，占比：{1:.2%}'.format(other_expendses, 
                                          other_expendses / total_expenses))
````
