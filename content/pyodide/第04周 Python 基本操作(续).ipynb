`````python
import os
#
fileNameList = [item for item in os.listdir('Data/Sales/') if item.endswith('.csv') and not item.startswith('.')]
saleAmt = 0
for fileName in fileNameList:  #
    saleFile = open('Data/Sales/' + fileName, encoding='utf8')
    saleFile.readline()  #
    for line in saleFile:  #
        detail = line.replace('"', '').strip('\n').split(',')
        if int(detail[5])>0:  #0
            saleAmt = saleAmt + (int(detail[4]) - int(detail[5])) * float(detail[3]) 
print('{:.2f}'.format(saleAmt))
`````

<font color=#A93226><b></b></font>  <font size=3 face="-">

Data/Sales</font>


`````python
import os
fileNameList = [item for item in os.listdir('Data/Sales/') if item.endswith('.csv') and item.startswith('2020')]
price_sum = 0
qty = 0
for fileName in fileNameList:  
    saleFile = open('Data/Sales/' + fileName, encoding='utf8')
    saleFile.readline()  
    for line in saleFile:  
        detail = line.replace('"', '').strip('\n').split(',')
        price_sum += float(detail[3])
        qty += 1
                           
print('{:.2f}'.format(price_sum/qty))
`````
